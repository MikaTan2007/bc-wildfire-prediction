### Joining both datasets

```python
import pandas as pd

df_negatives = pd.read_csv("bc_weather_negatives.csv")
df_positives = pd.read_csv("bc_weather_fires.csv")
```

```python
final_ml_dataset = pd.concat([df_negatives, df_positives], ignore_index=True)
```

```python
final_ml_dataset = final_ml_dataset.drop(columns=["SIZE_HA"])
```

```python
final_ml_dataset = final_ml_dataset.sample(frac = 1).reset_index(drop = True)
```

```python
final_ml_dataset = final_ml_dataset.sample(frac=1).reset_index(drop=True)
```
